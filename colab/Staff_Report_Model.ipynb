```python
import pandas as pd
from datetime import datetime, timedelta
import random

# Define data
data = {
    'staffname': ['Raj', 'Sita', 'Shyam', 'Rohan', 'Mohan', 'Sodi', 'Lalu', 'Hema', 'Hemang', 'Hemangi'],
    'staffusername': [147, 876, 765, 432, 321, 456, 987, 678, 123, 345],
    'timeappointed': [[] for _ in range(10)],
    'timeuploaded': [[] for _ in range(10)],
    'status': [[] for _ in range(10)]
}

# Generate random data for timeappointed, timeuploaded, and status for each user
for i in range(10):
    for _ in range(3):
        time_appointed = datetime.now() - timedelta(hours=random.randint(1, 24))
        time_uploaded = time_appointed + timedelta(hours=random.randint(1, 24))
        status = random.choice([0, 1])

        data['timeappointed'][i].append(time_appointed)
        data['timeuploaded'][i].append(time_uploaded)
        data['status'][i].append(status)

# Create DataFrame
df = pd.DataFrame(data)

# Display DataFrame
print(df)
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
import matplotlib.pyplot as plt

# Generate random data for demonstration
data = {
    'staffname': ['Raj', 'Sita', 'Shyam', 'Rohan', 'Mohan', 'Sodi', 'Lalu', 'Hema', 'Hemang', 'Hemangi'],
    'staffusername': [147, 876, 765, 432, 321, 456, 987, 678, 123, 345],
    'timeappointed': [[] for _ in range(10)],
    'timeuploaded': [[] for _ in range(10)],
    'status': [[] for _ in range(10)]
}

for i in range(10):
    for _ in range(3):
        time_appointed = datetime.now() - timedelta(hours=random.randint(1, 24))
        time_uploaded = time_appointed + timedelta(hours=random.randint(1, 24))
        status = random.choice([0, 1])

        data['timeappointed'][i].append(time_appointed)
        data['timeuploaded'][i].append(time_uploaded)
        data['status'][i].append(status)

# Create DataFrame
df = pd.DataFrame(data)
df
```

```python

# Calculate punctuality for each worker and scale it to a range of 0 to 5
df['punctuality'] = df.apply(lambda x: (x['timeuploaded'][0] - x['timeappointed'][0]).total_seconds() / 60, axis=1)
min_punctuality = df['punctuality'].min()
max_punctuality = df['punctuality'].max()
df['scaled_punctuality'] = df['punctuality'].apply(lambda x: ((x - min_punctuality) / (max_punctuality - min_punctuality)) * 5)

# Count the number of times the room was found clean and flagged for each worker
df['clean_count'] = df['status'].apply(lambda x: x.count(1))
df['flagged_count'] = df['status'].apply(lambda x: x.count(0))

# Generate report
report = df[['staffname', 'staffusername', 'scaled_punctuality', 'clean_count', 'flagged_count']]

# Plot pie chart for cleanliness flag count
plt.figure(figsize=(8, 6))
plt.pie(df['clean_count'], labels=df['staffname'], autopct='%1.1f%%', startangle=140)
plt.title('Cleanliness Flag Count for Each Worker')
plt.axis('equal')
plt.show()

print(report)
```
